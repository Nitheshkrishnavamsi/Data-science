```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='YearsExperience', y='Salary', data=df)
plt.title('Scatter Plot of YearsExperience vs. Salary')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

```python
# Regression plot
plt.figure(figsize=(10, 6))
sns.regplot(x='YearsExperience', y='Salary', data=df)
plt.title('Regression Plot of YearsExperience vs. Salary')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

```python
y=df["Salary"]
y
```

```python
x=df[["YearsExperience"]]
x
```

```python
#spliting the train and test data
x_train,x_test,y_train,y_test=train_test_split(x, y,test_size=0.30)
le=LinearRegression()
le.fit(x_train,y_train)
```

```python
y_pred=le.predict(x_test)
y_pred
```

```python
mse=mean_squared_error(y_test, y_pred)
mse
```

```python
print("mean squared error:",mse.round(3))
print("ROOT MEAN SQUARE ERROR",np.sqrt(mse).round(3))
print("R2 score:",r2_score(y_test, y_pred).round(3))
```

```python
#visualize the linear line
plt.scatter(x_test, y_test,color="Red")
plt.plot(x_test,y_pred,color="blue")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Regression Line")
plt.show()
```
