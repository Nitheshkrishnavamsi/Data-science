```python
import pandas as pd
# Load the dataset
d1 = pd.read_csv("C:/Users/kaart/Downloads/AmazonReviewsExtractor_10_reviews_2023-12-04 (1).csv",encoding='utf-8')
d1
```

```python
d1=d1.drop(["ID","Username","ProfileURL","Title"],axis=1)
d1
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```python
import string
# Remove punctuation from the text
translator = str.maketrans("", "", string.punctuation)
d1['Review Content'] = d1['Review Content'].apply(lambda text: text.translate(translator))

```

```python
d1['Review Content'] 
```

```python
# Perform sentiment analysis on each text and create a new column for sentiment scores
d1['sentiment_score'] = d1['Review Content'].apply(lambda text: sia.polarity_scores(text))
```

```python
d1['sentiment_score']
```

```python
d1
```

```python
# Assume d1["sentiment_score"] contains the dictionary as mentioned above
d1['neg_score'] = d1['sentiment_score'].apply(lambda x: x['neg'])
d1['neu_score'] = d1['sentiment_score'].apply(lambda x: x['neu'])
d1['pos_score'] = d1['sentiment_score'].apply(lambda x: x['pos'])
d1['compound_score'] = d1['sentiment_score'].apply(lambda x: x['compound'])
```

```python
d1
```

```python
# Classify sentiments into categories (positive, negative, neutral)
d1['sentiment_label'] = d1['sentiment_score'].apply(lambda scores: 'positive' if scores['compound'] > 0 else 'negative' if scores['compound'] < 0 else 'neutral')
d1
```

```python
# Split the data for training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(d1['Review Content'], d1['sentiment_label'], test_size=0.3, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# Vectorize the text data
vectorizer = CountVectorizer(stop_words='english')
X_train_vect = vectorizer.fit_transform(X_train)
X_test_vect = vectorizer.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
# Initialize the Logistic Regression model
logreg = LogisticRegression()
```

```python
# Train the model
logreg.fit(X_train_vect, y_train)
```

```python
# Predictions on training set
y_train_pred = logreg.predict(X_train_vect)
# Predictions on test set
y_test_pred = logreg.predict(X_test_vect)
```

```python
from sklearn.metrics import accuracy_score
# Accuracy calculation
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
```

```python
# Display the results
print("Training Accuracy:", train_accuracy)
print("Test Accuracy:", test_accuracy)
```
